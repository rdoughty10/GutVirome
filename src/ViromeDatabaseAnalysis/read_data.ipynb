```python
viral = split[split[0] =='Viruses']
viral_ranks = ranks[split[0] == 'Viruses']
viral_reads = len(viral)
viral
```

```python
assigned_tax_col = viral.apply(pd.Series.last_valid_index, axis=1)
print(assigned_tax_col)
assigned_virus = viral.lookup(assigned_tax_col.index, assigned_tax_col.values)
counts, virus = np.unique(assigned_virus, return_counts=True)
unique_viruses = len(virus)
virus_counts = pd.DataFrame(virus, counts, columns=['Counts'])
virus_counts = virus_counts.sort_values(by='Counts', ascending=False)
print(f'Number of unique assignments: {unique_viruses}')
print(virus_counts)

```

```python
def tax_counts(viral_ranks, viral, level='species'):
    species_cols = []
    for index, row in viral_ranks.iterrows():
        try:
            col = list(row).index(level)
        except:
            col = None
        species_cols.append([index, col])
    species_cols = pd.DataFrame(np.array(species_cols), columns=['index', 'col']).dropna().set_index('index').squeeze()
    assigned_virus = viral.lookup(species_cols.index, species_cols.values)
    virus, counts = np.unique(assigned_virus, return_counts=True)
    unique_viruses = len(virus)
    virus_counts = pd.DataFrame(counts, virus, columns=['Counts'])
    virus_counts = virus_counts.sort_values(by='Counts', ascending=False)
    print(f'Number of unique viral {level} assigned: {unique_viruses}')
    print(f'Total {level} assignments: {np.sum(counts)}')
    print(virus_counts)
    
tax_counts(viral_ranks, viral, 'species')
```

```python
tax_counts(viral_ranks, viral, 'genus')
```

```python
tax_counts(viral_ranks, viral, 'family')
```

```python
## plotting the number of hits per sseqid in mgv database
data = blastMGV.groupby(['sseqid'])['sseqid'].count()
reads_data = []
num_contigs = []
reads_cutoff = [0, 10, 20, 50, 100, 200, 500, 1000]

for cutoff in reads_cutoff:
    reads_data.append(data[data>cutoff])
    num_contigs.append(len(data[data>cutoff]))
    
labels = [f'{cutoff}\nn={num_contig}' for cutoff, num_contig in zip(reads_cutoff, num_contigs)]

plt.boxplot(reads_data,labels=labels)
plt.yscale('log')
plt.xlabel('\nCutoff for number of reads \n (n=number of MGV contigs above threshold)')
plt.ylabel('\nNumber of reads assigned to each contig')
plt.title('Number of reads per MGV contig at different read thresholds')
```

```python
## testing different thresholds
# pident [60, 70, 80, 90, 95, 99]
pident_thresholds = [0, 50, 60, 70, 80, 90, 95, 99]
new_reads_with_hits = []
number_unique_contigs = []

for threshold in pident_thresholds:
    filtered = mgv_with_info[mgv_with_info['pident'] > threshold]
    new_reads_with_hits.append(len(np.unique(filtered['qseqid'])))
    number_unique_contigs.append(len(np.unique(filtered['sseqid'])))
    
print(new_reads_with_hits)
print(number_unique_contigs)
```

```python
width = 0.4
fig, ax = plt.subplots()
x = np.arange(len(pident_thresholds))
ax.bar(x-width/2, new_reads_with_hits, color='r', width=width)
ax.bar(x+width/2, number_unique_contigs, color='b', width=width)
ax.set_xticklabels(pident_thresholds)
ax.set_title("Number of newly mapped reads and unique \ncontigs at different percentage of identical matches")
ax.set_ylabel("Number of newly mapped reads")
ax.set_xlabel("Percentage of Identical Matches Threshold")
```
